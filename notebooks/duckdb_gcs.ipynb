```python
import os
import duckdb

from dotenv import load_dotenv
```

```python
con = duckdb.connect()
```

```python
con.execute("INSTALL httpfs")
con.execute("LOAD httpfs")
```

```python
load_dotenv()
```

```python
access_id = os.environ["GCS_ACCESS_KEY_ID"]
secret = os.environ["GCS_SECRET_ACCESS_KEY"]
```

```python
con.execute("DROP SECRET gcs_secret")
```

```python
con.execute(f"""
    CREATE SECRET gcs_secret (
        TYPE s3,
        KEY_ID '{access_id}',
        SECRET '{secret}',
        ENDPOINT 'storage.googleapis.com',
        REGION 'auto',
        SCOPE 'gs://'
    )
""")
```

```python
con.execute("""
    COPY (SELECT 1 AS test_col) TO 'gs://pet-food-analytics-bucket/test/test.parquet' (FORMAT PARQUET)
""")
```

```python
result = con.execute("SELECT * FROM 'gs://pet-food-analytics-bucket/test/test.parquet'").fetchall()
print(result)
```
